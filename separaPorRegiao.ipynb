```python
import pandas as pd
lista = []
for i in range(1, 17):
    lista.append(i)
lista.remove(11)
lista.remove(12)
lista.remove(14)
```

```python
df = pd.read_excel("ifoodRioPreto.xlsx", skiprows=0, usecols=lista)
newDf = df.sort_values(by=['Bairro'], ignore_index=True)
newDf.sort_index()
display(newDf)
```

```python
lista = []
for bairro in df['Bairro']:
    if lista.count(bairro) == 0:
        lista.append(bairro)

k = len(df['Bairro'])
n = len(lista)

for i in range(0, n):
    lista[i] = lista[i].title().strip(' ')

for i in range(0,k):
    newDf['Bairro'][i] = newDf['Bairro'][i].title().strip(' ')
    
```

```python
n = pd.read_excel("DadosQuantitativos.xlsx", sheet_name="Regiões", usecols=[0], nrows = 60)
s = pd.read_excel("DadosQuantitativos.xlsx", sheet_name="Regiões", usecols=[1], nrows = 41)
l = pd.read_excel("DadosQuantitativos.xlsx", sheet_name="Regiões", usecols=[2], nrows = 71)
o = pd.read_excel("DadosQuantitativos.xlsx", sheet_name="Regiões", usecols=[3], nrows = 35)
c = pd.read_excel("DadosQuantitativos.xlsx", sheet_name="Regiões", usecols=[4], nrows = 17)

norte = []
sul = []
leste = []
oeste = []
central = []

#display(n)

for bairro in n['Zona Norte']:
    norte.append(bairro.title().strip(' '))    
for bairro in s['Zona Sul']:
    sul.append(bairro.title().strip(' '))    
for bairro in l['Zona Leste']:
    leste.append(bairro.title().strip(' '))
for bairro in o['Zona Oeste']:
    oeste.append(bairro.title().strip(' '))    
for bairro in c['Central']:
    central.append(bairro.title().strip(' '))
```

```python
regioes = []
k = len(df['Bairro'])

for i in range(0, k):
    if norte.count(newDf['Bairro'][i]) > 0:
        regioes.append('Norte')
    elif sul.count(newDf['Bairro'][i]) > 0:
        regioes.append('Sul')
    elif leste.count(newDf['Bairro'][i]) > 0:
        regioes.append('Leste')
    elif oeste.count(newDf['Bairro'][i]) > 0:
        regioes.append('Oeste')
    elif central.count(newDf['Bairro'][i]) > 0:
        regioes.append('Central')
    else:
        regioes.append('Indefinido')
```

```python
newDf['Região'] = regioes
```
